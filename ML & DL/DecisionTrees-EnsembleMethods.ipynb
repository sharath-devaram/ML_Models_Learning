```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# DecisionTrees Implementation

# 1 CART: Classification And Regression Trees

```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
#Load the dataset
data = load_breast_cancer()
x = data.data
y = data.target
```

```python
#divide into train and test
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#load the model and fit x_train, y_train
dt_cart = DecisionTreeClassifier()
dt_cart.fit(x_train, y_train)
```

```python
y_pred = dt_cart.predict(x_test)
y_pred[:20]
```

```python
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy: ",accuracy)
```

```python
##Visualization
import graphviz
```

```python
dot_data = tree.export_graphviz(dt_cart,out_file=None,feature_names=data.feature_names,class_names=data.target_names,filled=True,rounded=True,special_characters=True)
```

```python
graph = graphviz.Source(dot_data)
##save and render the decision tree as a pdf or image
graph.render("decision_tree",view=True)
```

# ID3

```python
id3_model=DecisionTreeClassifier(criterion='entropy')
id3_model.fit(x_train,y_train)
y_pred=id3_model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy: ",accuracy)
```

```python
c45_model =DecisionTreeClassifier(criterion='entropy',splitter='best')
c45_model.fit(x_train,y_train)
y_pred=c45_model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy: ",accuracy)
```

# Random Forests

```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data =load_breast_cancer()
x= data.data
y = data.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy: ",accuracy)
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb_model = GradientBoostingClassifier(n_estimators = 100, learning_rate =0.1,random_state=42)
gb_model.fit(x_train, y_train)
```

```python
y_pred = gb_model.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy: ",accuracy)
```
